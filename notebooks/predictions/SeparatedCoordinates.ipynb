```python
df_41 = df[df['POLYLINE_LENGTH'] == 41]
df_41
```

```python
import timesfm
```

```python
df_41_lon = df_41.drop(columns=['ORIGIN_CALL', 'TAXI_ID', 'POLYLINE_LENGTH', 'LAT'])
df_41_lat = df_41.drop(columns=['ORIGIN_CALL', 'TAXI_ID', 'POLYLINE_LENGTH', 'LON'])
```

```python
tfm_lat = timesfm.TimesFm(
      hparams=timesfm.TimesFmHparams(
          backend="gpu",
          per_core_batch_size=32,
          horizon_len=10,
          num_layers=50,
          use_positional_embedding=False,
          context_len=2048,
      ),
      checkpoint=timesfm.TimesFmCheckpoint(
          huggingface_repo_id="google/timesfm-2.0-500m-pytorch"),
  )
```

```python
tfm_lon = timesfm.TimesFm(
      hparams=timesfm.TimesFmHparams(
          backend="gpu",
          per_core_batch_size=32,
          horizon_len=10,
          num_layers=50,
          use_positional_embedding=False,
          context_len=2048,
      ),
      checkpoint=timesfm.TimesFmCheckpoint(
          huggingface_repo_id="google/timesfm-2.0-500m-pytorch"),
  )
```

```python
df_41_lat = df_41_lat.sort_values(by=['TRIP_ID', 'TIMESTAMP'])
df_41_lat_test = df_41_lat.groupby('TRIP_ID', group_keys=False).tail(10)
df_41_lat_train = df_41_lat.drop(df_41_lat_test.index)
```

```python
df_41_lon = df_41_lon.sort_values(by=['TRIP_ID', 'TIMESTAMP'])
df_41_lon_test = df_41_lon.groupby('TRIP_ID', group_keys=False).tail(10)
df_41_lon_train = df_41_lon.drop(df_41_lat_test.index)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_lat = MinMaxScaler()
df_41_lat_train['LAT'] = scaler_lat.fit_transform(df_41_lat_train[['LAT']])

scaler_lon = MinMaxScaler()
df_41_lon_train['LON'] = scaler_lon.fit_transform(df_41_lon_train[['LON']])
```

```python
df_41_lat_train['TIMESTAMP'] = pd.to_datetime(df_41_lat_train['TIMESTAMP'], unit='s')
df_41_lon_train['TIMESTAMP'] = pd.to_datetime(df_41_lon_train['TIMESTAMP'], unit='s')
```

```python
df_41_lon_train = df_41_lon_train.rename(columns={'TIMESTAMP': 'ds', 'LON': 'y', 'TRIP_ID': 'unique_id'})
df_41_lat_train = df_41_lat_train.rename(columns={'TIMESTAMP': 'ds', 'LAT': 'y', 'TRIP_ID': 'unique_id'})
```

```python
df_41_lat_train
```

```python
df_41_lon_train
```

```python
forecast_df_lon = tfm_lon.forecast_on_df(
    inputs=df_41_lon_train,
    freq="15s",
    value_name="y",
    num_jobs=-1,
)
```

```python
forecast_df_lat = tfm_lon.forecast_on_df(
    inputs=df_41_lat_train,
    freq="15s",
    value_name="y",
    num_jobs=-1,
)
```

```python
forecast_df_lon_keep = forecast_df_lon[['unique_id', 'ds', 'timesfm-q-0.9']]
forecast_df_lon_keep
```

```python
forecast_df_lat_keep = forecast_df_lat[['unique_id', 'ds', 'timesfm-q-0.9']]
forecast_df_lat_keep
```

```python
forecast_df_lat_keep['timesfm-q-0.9'] = scaler_lat.inverse_transform(forecast_df_lat_keep[['timesfm-q-0.9']])
forecast_df_lon_keep['timesfm-q-0.9'] = scaler_lon.inverse_transform(forecast_df_lon_keep[['timesfm-q-0.9']])
```

```python
forecast_df_lon_keep
```

```python
forecast_df_lat_keep
```

```python
forecast_df_lat_keep = forecast_df_lat_keep.rename(columns={'timesfm-q-0.9': 'LAT_PREDICTED'})
forecast_df_lon_keep = forecast_df_lon_keep.rename(columns={'timesfm-q-0.9': 'LON_PREDICTED'})
```

```python
forecast_df_lat_keep = forecast_df_lat_keep.sort_values(['unique_id', 'ds'])
forecast_df_lat_keep = forecast_df_lat_keep.groupby('unique_id').tail(1)

forecast_df_lon_keep = forecast_df_lon_keep.sort_values(['unique_id', 'ds'])
forecast_df_lon_keep = forecast_df_lon_keep.groupby('unique_id').tail(1)
```

```python
df_41_lon_test
```

```python
df_41_lon_test = df_41_lon_test.rename(columns={'TIMESTAMP': 'ds', 'TRIP_ID': 'unique_id'})
df_41_lon_test = df_41_lon_test.sort_values(['unique_id', 'ds'])
df_41_lon_test = df_41_lon_test.groupby('unique_id').tail(1)

df_41_lat_test = df_41_lat_test.rename(columns={'TIMESTAMP': 'ds', 'TRIP_ID': 'unique_id'})
df_41_lat_test = df_41_lat_test.sort_values(['unique_id', 'ds'])
df_41_lat_test = df_41_lat_test.groupby('unique_id').tail(1)
```

```python
result_lat = df_41_lat_test.merge(forecast_df_lat_keep, how='left', on=['unique_id'])
result_lon = df_41_lon_test.merge(forecast_df_lon_keep, how='left', on=['unique_id'])
result = result_lat.merge(result_lon, how='left', on=['unique_id'])
result
```

```python
from evaluation_script import haversine_distance
```

```python
result['DISTANCE_DELTA_SEPARATED'] = result.apply(lambda row: haversine_distance(
    row['LAT'], row['LON'], row['LAT_PREDICTED'], row['LON_PREDICTED']), axis=1)
result['DISTANCE_DELTA_SEPARATED'].mean()
```
