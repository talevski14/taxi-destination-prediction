```python
import math
```

```python
def haversine_distance(lat1, lon1, lat2, lon2):
    # Returns distance in kilometers
    R_earth = 6371  # Earth's radius in km

    # Convert degrees to radians
    lat_diff = math.radians(abs(lat1 - lat2))
    lon_diff = math.radians(abs(lon1 - lon2))
    lat1_rad = math.radians(lat1)
    lat2_rad = math.radians(lat2)

    # Haversine formula
    a = math.sin(lat_diff / 2) ** 2 + math.cos(lat1_rad) * math.cos(lat2_rad) * math.sin(lon_diff / 2) ** 2
    d = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))
    distance = R_earth * d

    return distance
```

```python
lat_original = float(df_processing_lat_test.tail(1)['y'])
lon_original = float(df_processing_long_test.tail(1)['y'])
lat_predicted = float(forecast_df_lat.tail(1)['timesfm'])
lon_predicted = float(forecast_df.tail(1)['timesfm'])

print(f'Latitude  - Real: {lat_original}, Predicted: {lat_predicted}')
print(f'Longitude - Real: {lon_original}, Predicted: {lon_predicted}')

distance = haversine_distance(lat_original, lon_original, lat_predicted, lon_predicted)
print(f'The distance between the predicted and the original ending point is {distance} km. - ON Q-0.5')
```
