```python
import math
```

```python
def haversine_distance(lat1, lon1, lat2, lon2):
    # Returns distance in kilometers
    R_earth = 6371  # Earth's radius in km

    # Convert degrees to radians
    lat_diff = math.radians(abs(lat1 - lat2))
    lon_diff = math.radians(abs(lon1 - lon2))
    lat1_rad = math.radians(lat1)
    lat2_rad = math.radians(lat2)

    # Haversine formula
    a = math.sin(lat_diff / 2) ** 2 + math.cos(lat1_rad) * math.cos(lat2_rad) * math.sin(lon_diff / 2) ** 2
    d = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))
    distance = R_earth * d

    return distance
```

```python
lat_original = float(df_processing_lat_test.tail(1)['y'])
lon_original = float(df_processing_long_test.tail(1)['y'])

lat_predicted = [float(forecast_df_lat.tail(1)[column]) for column in columns]
lon_predicted = [float(forecast_df_long.tail(1)[column]) for column in columns]

distances = []
for index, (lat, lon) in enumerate(zip(lat_predicted, lon_predicted)):
    index = index+1
    print(f'Q-0.{index}')
    print(f'Latitude  - Real: {lat_original}, Predicted: {lat}')
    print(f'Longitude - Real: {lon_original}, Predicted: {lon}')
    distance = haversine_distance(lat_original, lon_original, lat, lon)
    distances.append(distance)
    print(f'Distance: {distance} - Q-0.{index}')
```

```python
shortest_distance = min(distances)
shortest_distance_index = distances.index(shortest_distance) + 1

print(f'Shortest distance appears in Q-0.{shortest_distance_index}, with a distance of {shortest_distance} kilometers')
```
